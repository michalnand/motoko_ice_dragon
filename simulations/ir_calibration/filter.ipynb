```python
import numpy
import matplotlib.pyplot as plt




f       = 2.0*numpy.pi/4.0
r       = 0.9


a1 = -2.0*r*numpy.cos(f)
a2 = r**2

b0 = (1 - (r**2))/2.0
b1 = 0.0
b2 = -b0

print("b0 = ", round(b0, 6))
print("b1 = ", round(b1, 6))
print("b2 = ", round(b2, 6))
print("a1 = ", round(a1, 6))
print("a2 = ", round(a2, 6))


n_steps = 10000
w       = 2.0*numpy.pi*numpy.arange(0, n_steps)/n_steps + 10**-8

z = numpy.exp(w*1j)


h = (b0 + b1*(z**-1) + b2*(z**-2))/(1.0 + a1*(z**-1) + a2*(z**-2))

amp = numpy.absolute(h)


plt.plot(w/(2*numpy.pi), amp)
plt.show()
```
